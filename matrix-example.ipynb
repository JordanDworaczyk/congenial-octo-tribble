## Using the Convolution Matrix Operator

```python
import convolution.kernel
import convolution.matrix

from signals import rectangle
from domain import equispaced_points
```

```python
import matplotlib.pyplot as plt
```

```python
t = equispaced_points(512)
```

```python
latent_signal = rectangle(t)
```

```python
K = convolution.matrix.Operator(
    kernel=convolution.kernel.Gaussian(t, 20),
    structure=convolution.matrix.toeplitz
)
```

```python
plt.suptitle(f'Convolution Matrix from {K.kernel.__name__} Kernel')
plt.title(f'with {K.structure.__name__} structure and evaulated at ' + r'$\sigma = $' + f'{K.kernel.std}', fontsize=9)
plt.imshow(K.image)
plt.ylabel('Rows')
plt.xlabel('Columns')
plt.colorbar(label='Density');
```

```python
plt.suptitle('Derivative of Convolution Matrix')
plt.title(f'with w.r.t. standard deviation evaulated at ' + r'$\sigma = $' f'{K.kernel.std}', fontsize=9)
plt.imshow(K.partial_derivatives[0])
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.colorbar(label='Density');
```
